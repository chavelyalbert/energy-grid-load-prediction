```python

import reverse_geocode

df = spark.table("workspace.european_weather_raw.weather_hourly")
coordinates = df.select("lat", "lon").distinct().collect()
coord_list = [(row['lat'], row['lon']) for row in coordinates]
countries = [loc['country_code'] for loc in reverse_geocode.search(coord_list)]
result_df = spark.createDataFrame(list(zip(coord_list, countries)), ["coordinates", "country_code"])
display(result_df)
```
