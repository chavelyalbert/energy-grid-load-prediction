```python
generation = spark.table("workspace.default.train_set")
```

```python
display(generation.limit(5))
```

```python
from pyspark.sql.functions import dayofweek
from pyspark.sql.functions import date_format
from pyspark.sql.functions import hour, when, col
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
generation = generation.withColumn("day_of_week", dayofweek("Index")) #1-sunday
display(generation.limit(5))
```

```python
generation = generation.withColumn("hour", hour(col("Index")))

generation = generation.withColumn(
    "time_bin",
    when((col("hour") >= 6) & (col("hour") < 12), 1)
    .when((col("hour") >= 12) & (col("hour") < 18), 2)
    .when((col("hour") >= 18) & (col("hour") < 24), 3)
    .otherwise(4)   # covers 0â€“6
)

generation = generation.drop("hour")
display(generation.limit(5))
```

```python
generation_p= generation.toPandas()
numeric_pdf = generation_p.select_dtypes(include=[np.number])
corr_matrix = numeric_pdf.corr()

plt.figure(figsize=(10, 8))
plt.figure(figsize=(25, 20))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
#sns.clustermap(corr_matrix, figsize=(20, 20), cmap="coolwarm") -hierarchical clustering, we cannot do right now because of the null values
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.show()
# plt.savefig("/Workspace/Users/sokotushchenko@gmail.com/cor_plot.png") 

```
