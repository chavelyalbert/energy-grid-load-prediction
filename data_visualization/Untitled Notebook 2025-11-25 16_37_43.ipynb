```python
train_set = spark.table("workspace.default.train_set")
```

```python
df_train = train_set.toPandas()
df_train.head()
df_train.describe()
df_train.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

selected = ['ES', 'FR', 'DE', 'PT', 'IT']
fig, ax = plt.subplots()

sns.lineplot(x='index', y='grid_stress_score', hue='country', data=df_train[df_train['country'].isin(selected)], ax=ax)
ax.set(title='Grid Stress Score (Hourly)')

plt.show()
```
