```python


# Rename flow column
flows = crossborder.withColumnRenamed("_0", "flow_mw")
# Compute imports
imports = (flows
           .groupBy("to_country")
           .sum("flow_mw")
           .withColumnRenamed("to_country", "country")
           .withColumnRenamed("sum(flow_mw)", "import_mw"))

# Compute exports
exports = (flows
           .groupBy("from_country")
           .sum("flow_mw")
           .withColumnRenamed("from_country", "country")
           .withColumnRenamed("sum(flow_mw)", "export_mw"))

# Join both into your main df
df = df.join(imports, on="country", how="outer")
df = df.join(exports, on="country", how="outer")
```

```python
from pyspark.sql.functions import col, sum, when

# total rows
total_rows = generation.count()

# compute null count + missing pct for every column
summary = []

for c in generation.columns:
    null_count = generation.filter(col(c).isNull()).count()
    missing_pct = (null_count / total_rows) * 100
    summary.append((c, null_count, missing_pct))

# convert to Spark DataFrame for display
summary_df = spark.createDataFrame(summary, ["column", "null_count", "missing_pct"])

# show nicely
display(summary_df.orderBy("missing_pct", ascending=False))
```
